```python
# TUGAS
class TowerOfHanoi:
    def __init__(self, jumlah_cakram=3, langkah_maks=13):
        self.jumlah_cakram = jumlah_cakram
        self.langkah_maks = langkah_maks
        self.langkah = 0

        # Inisialisasi 3 tiang menggunakan Stack (list)
        self.tiang = {
            'A': list(range(jumlah_cakram, 0, -1)),  # Tiang awal dengan cakram terbesar di bawah
            'B': [],  # Tiang kosong
            'C': []   # Tiang tujuan
        }

    def pindah_cakram(self, dari_tiang, ke_tiang):
        """Memindahkan cakram dari satu tiang ke tiang lain."""
        if self.langkah >= self.langkah_maks:
            print("Anda telah mencapai batas maksimal langkah!")
            return

        if not self.tiang[dari_tiang]:
            print(f"Tiang {dari_tiang} kosong!")
            return

        if self.tiang[ke_tiang] and self.tiang[ke_tiang][-1] < self.tiang[dari_tiang][-1]:
            print("Tidak bisa menaruh cakram besar di atas yang lebih kecil!")
            return

        # Pindahkan cakram
        cakram = self.tiang[dari_tiang].pop()
        self.tiang[ke_tiang].append(cakram)
        self.langkah += 1
        print(f"Memindahkan cakram {cakram} dari {dari_tiang} ke {ke_tiang}")

    def tampilkan_tiang(self):
        """Menampilkan kondisi tiang-tiang saat ini."""
        for tiang, cakram in self.tiang.items():
            print(f"{tiang}: {cakram}")
        print(f"Langkah: {self.langkah}/{self.langkah_maks}\n")

    def selesai(self):
        """Memeriksa apakah permainan telah selesai."""
        return len(self.tiang['C']) == self.jumlah_cakram

# Permainan Interaktif
print("Selamat datang di Tower of Hanoi!")
game = TowerOfHanoi()

while not game.selesai():
    game.tampilkan_tiang()
    dari_tiang = input("Pilih tiang asal (A, B, C): ").upper()
    ke_tiang = input("Pilih tiang tujuan (A, B, C): ").upper()

    if dari_tiang in game.tiang and ke_tiang in game.tiang:
        game.pindah_cakram(dari_tiang, ke_tiang)
    else:
        print("Input tidak valid, silakan coba lagi.")

    if game.langkah >= game.langkah_maks:
        print("Anda telah mencapai batas maksimal langkah! Permainan berakhir.")
        break

if game.selesai():
    print("Selamat! Anda menyelesaikan permainan!")
```
