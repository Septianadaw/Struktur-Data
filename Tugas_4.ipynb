```python
# Latihan dan Tugas 1
# Implementasi sistem antrian pemesanan tiket bioskop
class TicketBioskop:
    def __init__(self, kapasitas=10):
        self.queue = []
        self.kapasitas = kapasitas

    def enqueue(self, ticket_id):
        """Menambahkan pelanggan ke dalam antrian jika masih ada tempat."""
        if len(self.queue) < self.kapasitas:
            self.queue.append(ticket_id)
            print(f"{ticket_id} telah masuk dalam antrian.")
        else:
            print("Antrian sudah penuh. Harap tunggu hingga ada tempat kosong.")

    def dequeue(self):
        """Melayani pelanggan pertama dalam antrian dan mengeluarkannya dari antrian."""
        if self.is_empty():
            print("Antrian kosong. Tidak ada pelanggan yang bisa dilayani.")
        else:
            pelanggan_dilayani = self.queue.pop(0)
            print(f"Sedang melayani pelanggan {pelanggan_dilayani}.")

    def front(self):
        """Menampilkan pelanggan pertama dalam antrian atau memberi tahu jika antrian kosong."""
        return self.queue[0] if not self.is_empty() else "Tidak ada pelanggan dalam antrian."

    def rear(self):
        """Menampilkan pelanggan terakhir dalam antrian atau memberi tahu jika antrian kosong."""
        return self.queue[-1] if not self.is_empty() else "Tidak ada pelanggan dalam antrian."

    def is_empty(self):
        """Memeriksa apakah antrian kosong."""
        return len(self.queue) == 0

    def size(self):
        """Menghitung jumlah pelanggan yang sedang mengantri."""
        return len(self.queue)

    def add_vip(self, ticket_id):
        """Memeriksa apakah pelanggan memiliki kode VIP dan langsung mendapatkan kursi."""
        if '8' in ticket_id:
            print(f"{ticket_id} adalah pelanggan VIP. Mendapatkan kursi tanpa perlu mengantri.")
        else:
            self.enqueue(ticket_id)

# Contoh Penggunaan
queue = TicketBioskop(kapasitas=10)
customers = ["SE131", "SE134", "SE136", "SE137", "SE137", "SE18", "SE139", "SE1310", "SE1311", "SE1312", "SE1314", "SE132", "SE1341", "SE1348"]

for customer in customers:
    queue.add_vip(customer)

print("\nAntrian saat ini:", queue.queue)
queue.dequeue()
print("Antrian setelah satu pelanggan dilayani:", queue.queue)
print("Pelanggan pertama dalam antrian:", queue.front())
print("Pelanggan terakhir dalam antrian:", queue.rear())
print("Jumlah pelanggan dalam antrian:", queue.size())
```

```python
# Latihan dan Tugas 2
class PhotocopySystem:
    def __init__(self):
        self.queue = []

    def enqueue(self, order_id, num_pages):
        """Menambahkan pesanan fotokopi ke dalam antrian."""
        if num_pages < 10:
            print(f"Pesanan {order_id} (Fast Track - {num_pages} halaman) langsung diproses.")
        else:
            self.queue.append((order_id, num_pages))
            print(f"Pesanan {order_id} ({num_pages} halaman) telah masuk dalam antrian.")

    def dequeue(self):
        """Melayani pesanan pertama dalam antrian."""
        if self.is_empty():
            print("Antrian kosong. Tidak ada pesanan yang bisa dilayani.")
        else:
            order_id, num_pages = self.queue.pop(0)
            if num_pages > 100:
                print(f"Pesanan {order_id} ({num_pages} halaman) diproses dalam dua tahap.")
                self.queue.append((order_id, num_pages - 50))
                print(f"Tahap pertama selesai. {order_id} tetap dalam antrian untuk tahap kedua.")
            else:
                print(f"Pesanan {order_id} ({num_pages} halaman) sedang diproses.")

    def front(self):
        """Menampilkan pesanan pertama dalam antrian."""
        return self.queue[0] if not self.is_empty() else "Tidak ada pesanan dalam antrian."

    def rear(self):
        """Menampilkan pesanan terakhir dalam antrian."""
        return self.queue[-1] if not self.is_empty() else "Tidak ada pesanan dalam antrian."

    def is_empty(self):
        """Memeriksa apakah antrian kosong."""
        return len(self.queue) == 0

    def size(self):
        """Menghitung jumlah pesanan dalam antrian."""
        return len(self.queue)

# Contoh Penggunaan
queue = PhotocopySystem()
orders = [("SEP13", 120), ("SEP13", 5), ("SEP13", 30), ("SEP13", 150)]

for order in orders:
    queue.enqueue(order[0], order[1])

print("\nAntrian saat ini:", queue.queue)
queue.dequeue()
print("Antrian setelah satu pesanan dilayani:", queue.queue)
print("Pesanan pertama dalam antrian:", queue.front())
print("Pesanan terakhir dalam antrian:", queue.rear())
print("Jumlah pesanan dalam antrian:", queue.size())
```

```python
# Latihan dan Tugas 3
class MallParking:
    def __init__(self, kapasitas=15):
        self.vehicles = []
        self.kapasitas = kapasitas

    def enqueue(self, vehicle_number, is_electric=False):
        """Menambahkan kendaraan ke dalam antrian parkir."""
        if len(self.vehicles) >= self.kapasitas:
            print(f"Parkiran penuh. Kendaraan {vehicle_number} harus menunggu.")
        elif is_electric:
            self.vehicles.insert(0, vehicle_number)
            print(f"Kendaraan {vehicle_number} (Listrik - Parkir Prioritas) langsung masuk parkiran.")
        else:
            self.vehicles.append(vehicle_number)
            print(f"Kendaraan {vehicle_number} masuk ke antrian parkir.")

    def dequeue(self):
        """Mengeluarkan kendaraan pertama dari antrian parkir."""
        if self.is_empty():
            print("Parkiran kosong. Tidak ada kendaraan yang keluar.")
        else:
            removed_vehicle = self.vehicles.pop(0)
            print(f"Kendaraan {removed_vehicle} keluar dari parkiran.")

    def front(self):
        """Menampilkan kendaraan pertama dalam antrian."""
        return self.vehicles[0] if not self.is_empty() else "Tidak ada kendaraan dalam antrian."

    def rear(self):
        """Menampilkan kendaraan terakhir dalam antrian."""
        return self.vehicles[-1] if not self.is_empty() else "Tidak ada kendaraan dalam antrian."

    def is_empty(self):
        """Memeriksa apakah parkiran kosong."""
        return len(self.vehicles) == 0

    def size(self):
        """Menghitung jumlah kendaraan dalam antrian."""
        return len(self.vehicles)

# Contoh Penggunaan
parking = MallParking()
vehicles = [("KL13", False), ("BA24", False), ("SU57", True), ("MA91", False), ("BE89", True)]

for vehicle in vehicles:
    parking.enqueue(vehicle[0], vehicle[1])

print("\nAntrian parkiran saat ini:", parking.vehicles)
parking.dequeue()
print("Antrian setelah satu kendaraan keluar:", parking.vehicles)
print("Kendaraan pertama dalam antrian:", parking.front())
print("Kendaraan terakhir dalam antrian:", parking.rear())
print("Jumlah kendaraan dalam antrian:", parking.size())
```

```python
# Latihan dan Tugas 4
class AntrianSertifikat:
    def __init__(self):
        self.regular_queue = []  # Antrian mahasiswa reguler
        self.prestasi_queue = []  # Antrian mahasiswa berprestasi

    def enqueue(self, student_id, is_prestasi=False, is_nasional=False):
        """Menambahkan mahasiswa ke dalam antrian pengambilan sertifikat."""
        if is_nasional:
            print(f"Mahasiswa {student_id} langsung mengambil sertifikat tanpa antri.")
        elif is_prestasi:
            self.prestasi_queue.append(student_id)
            print(f"Mahasiswa {student_id} (Berprestasi) masuk ke antrian prioritas.")
        else:
            self.regular_queue.append(student_id)
            print(f"Mahasiswa {student_id} masuk ke antrian reguler.")

    def dequeue(self):
        """Melayani mahasiswa yang lebih dulu datang."""
        if self.prestasi_queue:
            removed_student = self.prestasi_queue.pop(0)
        elif self.regular_queue:
            removed_student = self.regular_queue.pop(0)
        else:
            print("Antrian kosong. Tidak ada mahasiswa yang dilayani.")
            return
        print(f"Mahasiswa {removed_student} telah mengambil sertifikat.")

    def front(self):
        """Menampilkan mahasiswa pertama dalam antrian."""
        if self.prestasi_queue:
            return self.prestasi_queue[0]
        elif self.regular_queue:
            return self.regular_queue[0]
        return "Tidak ada mahasiswa dalam antrian."

    def rear(self):
        """Menampilkan mahasiswa terakhir dalam antrian."""
        if self.regular_queue:
            return self.regular_queue[-1]
        elif self.prestasi_queue:
            return self.prestasi_queue[-1]
        return "Tidak ada mahasiswa dalam antrian."

    def is_empty(self):
        """Memeriksa apakah antrian kosong."""
        return not self.prestasi_queue and not self.regular_queue

    def size(self):
        """Menghitung jumlah mahasiswa dalam antrian."""
        return len(self.prestasi_queue) + len(self.regular_queue)

# Contoh Penggunaan
antrian = AntrianSertifikat()
mahasiswa = [("TI13", False, False), ("TI21", True, False), ("TI09", False, False), ("TI05", False, True)]

for mhs in mahasiswa:
    antrian.enqueue(mhs[0], mhs[1], mhs[2])

print("\nAntrian saat ini:")
print("Mahasiswa Berprestasi:", antrian.prestasi_queue)
print("Mahasiswa Reguler:", antrian.regular_queue)

antrian.dequeue()
print("Antrian setelah satu mahasiswa dilayani:")
print("Mahasiswa Berprestasi:", antrian.prestasi_queue)
print("Mahasiswa Reguler:", antrian.regular_queue)

print("Mahasiswa pertama dalam antrian:", antrian.front())
print("Mahasiswa terakhir dalam antrian:", antrian.rear())
print("Jumlah mahasiswa dalam antrian:", antrian.size())
```
